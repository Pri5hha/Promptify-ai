```python
!pip install transformers torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline

emotion_classifier = pipeline("text-classification",
                              model="bhadresh-savani/distilbert-base-uncased-emotion",
                              return_all_scores=True)
```

```python
def get_emotions(text):
    results = emotion_classifier(text)[0]  # list of dicts
    # Convert to dict {emotion: score}
    return {res['label']: res['score'] for res in results}
```

```python
text = "Buy this product now or you'll regret it!"
print(get_emotions(text))
```

```python
manipulation_keywords = {
    "fear": ["regret", "afraid", "danger", "loss"],
    "urgency": ["now", "hurry", "limited", "today"],
    "guilt": ["should", "must", "if you don't", "blame"],
    "exaggeration": ["always", "never", "everybody", "everyone"],
    "clickbait": ["shocking", "you won't believe", "secret", "revealed"]
}
```

```python
def highlight_text(text):
    highlighted = text
    for tactic, words in manipulation_keywords.items():
        for word in words:
            highlighted = highlighted.replace(word, f"[{word}]{tactic.upper()}")
    return highlighted
```

```python
print(highlight_text(text))
```

```python
def manipulation_score(zero_shot_result, emotion_scores):
    # Take top manipulation label score + sum of fear/guilt intensities
    top_score = zero_shot_result['scores'][0]
    fear_guilt = emotion_scores.get('fear', 0) + emotion_scores.get('guilt', 0)
    score = (top_score + fear_guilt) / 2 * 100
    return round(score, 2)
```

```python
zs_result = classifier(text, candidate_labels)
emotions = get_emotions(text)
print("Manipulation Score:", manipulation_score(zs_result, emotions))
```

```python
def plot_radar(zs_result, emotions):
    categories = zs_result['labels'][:5] + list(emotions.keys())[:5]
    values = zs_result['scores'][:5] + [emotions[e] for e in emotions.keys()][:5]

    fig = go.Figure()
    fig.add_trace(go.Scatterpolar(
        r=values,
        theta=categories,
        fill='toself',
        name='Manipulation Radar'
    ))
    fig.update_layout(polar=dict(radialaxis=dict(visible=True, range=[0,1])), showlegend=False)
    fig.show()
```

```python
def compute_depth_score(zs_result, emotion_scores, persuasion_score=0):
    top_manip_score = zs_result['scores'][0]
    fear_guilt_intensity = emotion_scores.get('fear',0) + emotion_scores.get('guilt',0)
    # optional: add persuasion_score if using keyword detection
    score = (top_manip_score + fear_guilt_intensity + persuasion_score) / 3 * 100
    return round(score, 2)
```

```python
import nltk
import os

# Define a directory to download NLTK data
nltk_data_dir = '/tmp/nltk_data'
os.makedirs(nltk_data_dir, exist_ok=True)

# Set the NLTK data path
nltk.data.path.append(nltk_data_dir)

# Download the punkt and punkt_tab resources to the specified directory
nltk.download('punkt', download_dir=nltk_data_dir)
nltk.download('punkt_tab', download_dir=nltk_data_dir)

from nltk import sent_tokenize

sentences = sent_tokenize(user_input_text)
print(sentences)
```

```python
results = []

for sent in sentences:
    zs_result = classifier(sent, candidate_labels)  # your existing zero-shot classifier
    results.append({'sentence': sent, 'zs_result': zs_result})
```

```python
for res in results:
    emotions = get_emotions(res['sentence'])  # your emotion detection function
    res['emotions'] = emotions
```

```python
def get_persuasion_scores(sentence):
    persuasion_keywords = {
        "bandwagon": ["everyone", "all", "join us", "don’t miss out"],
        "authority": ["experts say", "studies show", "scientists recommend"],
        "scarcity": ["limited", "only today", "hurry", "while stocks last"],
        "reciprocity": ["we give", "special offer", "bonus"]
    }
    scores = {}
    for tactic, keywords in persuasion_keywords.items():
        scores[tactic] = sum(1 for word in keywords if word.lower() in sentence.lower())
    # Normalize scores to 0-1
    max_count = max(scores.values()) if max(scores.values()) != 0 else 1
    scores = {k: v/max_count for k,v in scores.items()}
    return scores

for res in results:
    res['persuasion'] = get_persuasion_scores(res['sentence'])
```

```python
for res in results:
    zs = res['zs_result']
    emotions = res['emotions']
    persuasion_avg = sum(res['persuasion'].values()) / len(res['persuasion'])
    # Combine top zero-shot score, fear+guilt intensity, and persuasion
    top_score = zs['scores'][0]
    fear_guilt = emotions.get('fear',0) + emotions.get('guilt',0)
    depth_score = (top_score + fear_guilt + persuasion_avg) / 3 * 100
    res['depth_score'] = round(depth_score,2)
```

```python
top_sentences = sorted(results, key=lambda x: x['depth_score'], reverse=True)[:3]
print("Top manipulative sentences:")
for ts in top_sentences:
    print(f"{ts['sentence']} (Score: {ts['depth_score']})")
```

```python
import pandas as pd
import plotly.express as px

# Collect all keywords
all_keywords = []
for tactic, words in manipulation_keywords.items():
    all_keywords.extend(words)
all_keywords = list(set(all_keywords))

# Build intensity matrix
matrix = []
for res in results:
    row = []
    sentence = res['sentence'].lower()
    for kw in all_keywords:
        # Check if keyword in sentence
        if kw in sentence:
            # Intensity = depth score (normalized 0-1)
            row.append(res['depth_score']/100)
        else:
            row.append(0)
    matrix.append(row)

df = pd.DataFrame(matrix, columns=all_keywords, index=[res['sentence'] for res in results])
```

```python
!pip install ipywidgets matplotlib seaborn nltk

import ipywidgets as widgets
from IPython.display import display, clear_output
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
from nltk import sent_tokenize
import numpy as np

nltk.download('punkt')
nltk.download('punkt_tab')

# -----------------------------
# Manipulation keywords
# -----------------------------
manipulation_keywords = {
    "fear": ["regret", "afraid", "danger", "risk", "loss"],
    "urgency": ["now", "hurry", "limited", "today", "immediately"],
    "guilt": ["should", "must", "blame", "fault"],
    "exaggeration": ["always", "never", "everyone", "everybody"],
    "clickbait": ["shocking", "secret", "revealed", "won’t believe", "guarantee"]
}

# -----------------------------
# Persuasion tactics keywords
# -----------------------------
persuasion_keywords = {
    "authority": ["experts say", "studies show", "scientists recommend"],
    "scarcity": ["limited", "only today", "hurry", "while stocks last"],
    "reciprocity": ["we give", "special offer", "bonus"],
    "bandwagon": ["everyone", "all", "join us", "don’t miss out"]
}

# -----------------------------
# Emotion keywords
# -----------------------------
emotion_keywords = {
    "fear": ["regret", "afraid", "danger", "risk", "loss"],
    "guilt": ["should", "must", "blame", "fault"],
    "sadness": ["sad", "unhappy", "mourn", "sorrow"],
    "anger": ["angry", "furious", "rage", "hate"],
    "joy": ["happy", "joy", "delighted", "excited"]
}

# -----------------------------
# Suggest neutral version
# -----------------------------
def suggest_neutral(sentence):
    replacements = {
        "must": "should",
        "never": "rarely",
        "always": "often",
        "guarantee": "aim to",
        "limited": "few",
        "secret": "information",
        "hurry": "please act soon",
        "regret": "consider"
    }
    new_sentence = sentence
    for k, v in replacements.items():
        new_sentence = new_sentence.replace(k, f"{v}").replace(k.capitalize(), v.capitalize())
    return new_sentence

# -----------------------------
# Analyze function
# -----------------------------
def analyze_text(text):
    sentences = sent_tokenize(text)
    sentence_results = []
    category_totals = {k:0 for k in manipulation_keywords.keys()}
    persuasion_totals = {k:0 for k in persuasion_keywords.keys()}
    emotion_totals = {k:0 for k in emotion_keywords.keys()}

    for sent in sentences:
        sent_lower = sent.lower()
        # Manipulation counts
        category_counts = {k:0 for k in manipulation_keywords.keys()}
        words_found = []
        for cat, words in manipulation_keywords.items():
            for w in words:
                count = sent_lower.count(w)
                if count > 0:
                    category_counts[cat] += count
                    words_found.extend([w]*count)
        # Persuasion counts
        persuasion_counts = {k:0 for k in persuasion_keywords.keys()}
        for tactic, words in persuasion_keywords.items():
            for w in words:
                if w.lower() in sent_lower:
                    persuasion_counts[tactic] += 1
        # Emotion counts
        emotion_counts = {k:0 for k in emotion_keywords.keys()}
        for emo, words in emotion_keywords.items():
            for w in words:
                if w.lower() in sent_lower:
                    emotion_counts[emo] +=1
        # Depth score
        depth_score = min(sum(category_counts.values())*10, 100)
        sentence_results.append({
            "sentence": sent,
            "depth_score": depth_score,
            "categories": category_counts,
            "persuasion": persuasion_counts,
            "emotions": emotion_counts,
            "words_found": words_found
        })
        for k in category_counts: category_totals[k] += category_counts[k]
        for k in persuasion_counts: persuasion_totals[k] += persuasion_counts[k]
        for k in emotion_counts: emotion_totals[k] += emotion_counts[k]

    overall_score = min(sum(category_totals.values())*10, 100)
    return sentence_results, overall_score, category_totals, persuasion_totals, emotion_totals

# -----------------------------
# Widgets
# -----------------------------
text_box = widgets.Textarea(
    value="Paste text to analyze here...",
    placeholder='Enter text...',
    description='Input:',
    layout=widgets.Layout(width="100%", height="120px")
)

button = widgets.Button(description="Analyze", button_style="info")
output = widgets.Output()

# -----------------------------
# Button click function
# -----------------------------
def on_button_click(b):
    with output:
        clear_output()
        text = text_box.value
        sentence_results, overall_score, category_totals, persuasion_totals, emotion_totals = analyze_text(text)

        # -----------------------------
        # Display summary
        # -----------------------------
        print("=== Overall Manipulation Score ===")
        print(f"Score: {overall_score}/100")
        status = "⚠️ Manipulative text detected" if overall_score>0 else "✅ Neutral text"
        print(f"Status: {status}\n")

        # Highlight manipulative words
        highlighted_text = text
        for res in sentence_results:
            for w in set(res["words_found"]):
                highlighted_text = highlighted_text.replace(w, f"**{w.upper()}**")
        print("=== Highlighted Text ===")
        print(highlighted_text)

        # Neutral rewrite
        print("\n=== Suggested Neutral Version ===")
        print(suggest_neutral(text))

        # -----------------------------
        # Pie chart of manipulation categories
        # -----------------------------
        labels = [k.capitalize() for k,v in category_totals.items() if v>0]
        sizes = [v for v in category_totals.values() if v>0]
        if sizes:
            plt.figure(figsize=(6,6))
            plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=plt.cm.tab20.colors)
            plt.title("Manipulation Category Distribution")
            plt.show()

        # -----------------------------
        # Persuasion radar chart
        # -----------------------------
        labels = list(persuasion_totals.keys())
        values = [persuasion_totals[k] for k in labels]
        if sum(values)>0:
            angles = np.linspace(0, 2*np.pi, len(labels), endpoint=False).tolist()
            values += values[:1]  # close the circle
            angles += angles[:1]
            plt.figure(figsize=(6,6))
            ax = plt.subplot(111, polar=True)
            ax.plot(angles, values, 'o-', linewidth=2)
            ax.fill(angles, values, alpha=0.25)
            ax.set_thetagrids(np.degrees(angles[:-1]), labels)
            ax.set_title("Persuasion Tactics Radar")
            plt.show()

        # -----------------------------
        # Emotion bar chart
        # -----------------------------
        labels = list(emotion_totals.keys())
        values = [emotion_totals[k] for k in labels]
        if sum(values)>0:
            plt.figure(figsize=(8,4))
            plt.bar(labels, values, color='skyblue')
            plt.ylabel("Count")
            plt.title("Emotion Manipulation Counts")
            plt.show()

        # -----------------------------
        # Sentence-level heatmap
        # -----------------------------
        heatmap_data = []
        sentence_labels = []
        category_labels = list(manipulation_keywords.keys())
        for res in sentence_results:
            row = [res["categories"][cat] for cat in category_labels]
            heatmap_data.append(row)
            sentence_labels.append(res["sentence"][:50]+"..." if len(res["sentence"])>50 else res["sentence"])
        if any(any(row) for row in heatmap_data):
            plt.figure(figsize=(10, len(sentence_results)*0.5 + 2))
            sns.heatmap(heatmap_data, annot=True, fmt="d", cmap="Reds",
                        yticklabels=sentence_labels, xticklabels=[cat.capitalize() for cat in category_labels])
            plt.title("Sentence-level Manipulation Heatmap")
            plt.xlabel("Manipulation Categories")
            plt.ylabel("Sentences")
            plt.show()

button.on_click(on_button_click)
display(text_box, button, output)
```
